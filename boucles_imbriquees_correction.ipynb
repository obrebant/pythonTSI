# Exercice 4 : sommes doubles

## Un premier exemple pour comprendre

On considère la somme $\displaystyle S_1=\sum_{i=1}^n\sum_{j=1}^n ij$

**Q1. Partie papier sur feuille d'exercices :**

> **Q1a.** Calculer "à la main" S1 pour n=2

**Correction :** pour n=2 on trouve S1 = 9

> **Q1b** Calculer mathématiquement S1 en fonction de n

**Correction :** Pour n quelconque on trouve : $S_1=\left(\frac{n(n+1)}{2}\right)^2$

**Q2. Partie informatique :**

 > **Q2a.** Ecrire une fonction `S1(n : int)-> float` qui reçoit un entier n et qui renvoit la somme S1 correspondante

```python
#correction :
def S1(n : int) -> int:
    assert n>=1
    s = 0
    for i in range(1,n+1):
        for j in range(1,n+1):
            s += i*j
    return s
```

> **Q2b.** Ecrire une fonction `test_S1(n : int)->float` qui teste la validité de la fonction S1 pouir n=2 et
pour une dizaine de valeurs de n choisies aléatoirement entre 1 et 100

```python
#correction
from random import randint

def test_S1():
    assert S1(2)==9
    for _ in range(10):
        n = randint(1,100)
        assert S1(n)==(n*(n+1)/2)**2
    print("Tout est OK")

test_S1()
```

```python
#correction S2:
from random import randint

def S2(n : int) -> int:
    assert n>=1
    s = 0
    for i in range(1,n+1):
        for j in range(1,n+1):
            s += i+j
    return s

def test_S2():
    assert S2(2)==12
    for _ in range(10):
        n = randint(1,100)
        assert S2(n)==n**2*(n+1)
    print("Tout est OK")

test_S2()
```

# Exo 5 : triangle de Pascal

On rappelle la formule de Pascal permettant de calculer "de proche en proche" les coefficients binomiaux de son triangle. Pour $n\geqslant1$ et $0\leqslant p \leqslant n$, on a :
$$C(n+1,p)=C(n,p)+C(n,p-1)$$
 avec pour tout $n\geqslant 1$, $C(n,0)=C(n,n)=1$

**Q1** Ecrire à la main dans un tableau ($n$ en ligne, $p$ en colonne) les valeurs du triangle pour $n=1,...,4$ et vérifier la cohérence de la formule.

| | p=0 | p=1 | p=2 | p=3 | p=4 |
|---|---|---|---|---|---|
|n=1|1|1 | | | |
|n=2|1 | 2 | 1 |
|n=3|1 | 3 | 3 | 1 |
|n=4|1 | 4 | 6 | 4 | 1 |

**Q2** Ecrire une fonction python `C(n : int, p : int)` qui reçoit deux entiers n et p et qui renvoie le coefficient de Pascal correspondant. On n'oubliera pas d'écrire au préalable une fonction de test.

```python
def test_C():
    assert C(1,0)==1
    assert C(1,1)==1
    assert C(4,0)==1
    assert C(4,2)==6
    print("Tout est OK !")

def C(n : int,p : int) -> int:
    """ renvoie le coefficient de Pascal C(n,p)"""
    assert n>=1
    assert 0<=p<=n
    if p==0 or n==p :
        return 1
    return C(n-1,p)+C(n-1,p-1)

test_C()
```

**Q3** Ecrire une fonction python qui reçoit un entier n et qui affiche tout le triangle de Pascal jusq'à n.

```python
def triangle(n : int):
    """ affiche le triangle de Pascal de 1 jusqu'à n"""
    
    for i in range(1,n+1):
        lig =[]
        for p in range(i+1):
            lig.append(C(i,p))
        print(lig)

triangle(6)
```

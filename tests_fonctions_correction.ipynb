## Exercice 1 :

On suppose avoir trouvé un algorithme astucieux pour calculer le produit de deux entiers (c'est la multiplication que faisaient les égyptiens du temps des pharaons...). Cela donne une fonction `mult(x:int,y:int)->int` qui renvoie le produit de deux entiers passés en argument. 

```python
def mult(a,b):
    #multiplication egyptienne
    A = a
    B = b
    S = 0 
    while A != 0:
        if A % 2 == 1:
            S = S+B
        A = A//2
        B = B*2
        #print(A,B,S)
    return S
```

```python
mult(12,5)
```

```python
def test_mult():
    assert mult(0,2) == 0
    assert mult(2,1.5) == 3
    assert mult(10,1000) == 10000
    print("Tous les tests sont passés")
test_mult()
```

**Q2.b.** Proposer une correction de la fonction `mult` permettant de ne pas pouvoir tomber dans ces cas...

On ajoute des conditions avec "assert" pour s'assurer de la positivité des arguments :

```python
def mult(a,b):
    #multiplication egyptienne
    assert a>=0 # cet algo part en boucle infinie si les nombres sont négatifs
    assert b>=0
    A = a
    B = b
    S = 0 
    while A != 0:
        if A % 2 == 1:
            S = S+B
        A = A//2
        B = B*2
        #print(A,B,S)
    return S
```

##  Exercice 2

Imaginons que l'on ait écrit une fonction `tri(t: list)->list` dont le but est de trier le tableau en place (elle modifie directement le tableau t). On commence par écrire une fonction `test_tableau_trie(t:list)` qui vérifie que le résultat est correct (en utilisant la fonction `assert`), puis on appelle cette fonction sur plusieurs exemples bien choisis (aléatoirements !):

**Q1.** Pour tester si un tableau numérique est trié dans l'ordre croissant, on peu par exemple parcourir tous ses éléments et s'assurer qu'ils sont de plus en plus grands. Écrire une fonction `test_tableau_trie(t :list[float])` qui reçoit un tableau de flottants et qui *lève une exception* (c'est à dire renvoie une erreur) si ce tableau n'est pas trié dans l'ordre croissant.

```python
def test_tableau_trie(t):
    for i in range(len(t)-1):
        assert t[i]<=t[i+1]
```

**Q2.** Écrire uen fonction `tableau_aleatoire(n : int,a : int,b : int) -> list[int]` qui renvoit une liste de `n` entiers aléatoires compris entre `a`et `b` (inclus).

```python
from random import randint
def tableau_aleatoire(n,a,b):
    """ renvoie un tableau aleatoire de n elements
    choisis entre les entiers a et b"""
    tab = []
    for i in range(n):
        tab.append(randint(a,b))
    return tab
        
print(tableau_aleatoire(5,-3,4))
print(tableau_aleatoire(0,-10,10))
print(tableau_aleatoire(10,-100,100))
```

**Q3.** On peut maintenant tester notre fonction `tri` ci-dessous par le code suivant (tester pour voir si tout fonctionne)

```python
def tri(t):
    # on utilise la fonction prédéfinie python pour faire des essais
    # on pourra aussi utiliser votre fonction de tri à bulle faite avec M. Gras
    return sorted(t) 

#jeu de tests :

for n in range(100):
    t = tri(tableau_aleatoire(n,0,0)) #elements tous identiques
    test_tableau_trie(t)
    
    t = tri(tableau_aleatoire(n, -n//4, n//4)) #presence de doublons
    test_tableau_trie(t)
    
    t = tri(tableau_aleatoire(n, -10*n, 10*n)) #elements eparses
    test_tableau_trie(t)
    
print('tous les tests sont passés')
```

## Partie 2 : tests de "performances"

## Exercice 3 :

**Q1** Écrire un jeu de tests de performance pour la fonction `mult` de l'exercice 1 et comparer avec la multiplication native de python. Il faudra probablement tester la répétition d'un grand nombre de fois l'appel à la fonction...

```python
import time
lst1 = [0, 10, 50, 100]
lst2 = [15,30,40, 50]
t0 = time.perf_counter()
for i in range(10000):
    for a,b in zip(lst1,lst2) :
        mult(a,b)
t1 = time.perf_counter()
print("avec mult :",t1-t0)
for i in range(10000):
    for a,b in zip(lst1,lst2) :
        a*b
t2 = time.perf_counter()
print("avec * :",t2-t1)
```

**Q2** Améliorer ce jeu de tests en utilisant des nombres de plus en plus grands, par exemple des nombres de $2^0$ chiffres à $2^{15}$ chiffres en doublant la taille à chaque fois. Rappelons si besoin qu'un nombre à $n$ chiffres est compris dans l'intervalle $[10^{n};10^{n+1}-1]$.

```python
taille, temps1, temps2 = [], [], []
for k in range(3,8):
    # on teste des nombres a 2^k chiffres
    n=2**k
    mini = 10**n
    maxi = 10**(n+1)-1
    x = randint(mini,maxi)
    y = randint(mini,maxi)
    t0 = time.perf_counter()
    for i in range(10000): mult(x,y)
    t1 = time.perf_counter()
    for i in range(10000): x*y
    t2 = time.perf_counter()
    print(n,t1-t0,t2-t1)
    taille.append(n)
    temps1.append(t1-t0)
    temps2.append(t2-t1)
```

**Q3** Tracer la représentation graphique (avec le module `matplotlib.pyplot`) des temps d'exécution en fonction du nombre de chiffres.

```python
import matplotlib.pyplot as plt

plt.clf()
plt.figure(1)
plt.plot(taille,temps1)
plt.plot(taille,temps2)
plt.show()
```

```python
temps=[]
for t1,t2 in zip(temps1,temps2):
    temps.append(t1/t2)

plt.figure(2)
plt.plot(taille,temps)
plt.show()
```
